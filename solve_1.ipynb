```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import FontProperties
import platform
import warnings
warnings.filterwarnings('ignore')

# 设置中文字体
def set_chinese_font():
    system = platform.system()
    if system == 'Windows':
        plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'SimSun']
        font_path = 'C:/Windows/Fonts/simhei.ttf'''

    plt.rcParams['axes.unicode_minus'] = False

    try:
        chinese_font = FontProperties(fname=font_path)
        return chinese_font
    except:
        print(f"警告: 无法加载字体 {font_path}，将使用系统默认字体")
        return None

# 设置中文字体
chinese_font = set_chinese_font()

# 设置图表样式
plt.style.use('default')
plt.rcParams['axes.linewidth'] = 0.8
numeric_columns = [
    '年龄', '身高', '体重', '检测抽血次数', '检测孕周_周数', '孕妇BMI',
    '原始读段数', '在参考基因组上比对的比例', '重复读段的比例',
    '唯一比对的读段数', 'GC含量', '13号染色体的Z值', '18号染色体的Z值',
    '21号染色体的Z值', 'X染色体的Z值', 'Y染色体的Z值',
    'Y染色体浓度', 'X染色体浓度', '13号染色体的GC含量', '18号染色体的GC含量', '被过滤掉读段数的比例'
]
existing_columns = [col for col in numeric_columns if col in data_boy.columns]

for col in numeric_columns:
    if col in data_boy.columns:
        data_boy[col] = pd.to_numeric(data_boy[col], errors='coerce')


n_cols_per_fig = 4
n_figs = int(np.ceil(len(existing_columns) / n_cols_per_fig))


for fig_idx in range(n_figs):
    start_idx = fig_idx * n_cols_per_fig
    end_idx = min((fig_idx + 1) * n_cols_per_fig, len(existing_columns))
    current_columns = existing_columns[start_idx:end_idx]

    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    axes = axes.flatten()

    # 绘制每个变量的箱线图
    for i, col in enumerate(current_columns):
        ax = axes[i]
        data = data_boy[col].dropna()

        # 绘制箱线图
        boxplot = ax.boxplot(data, patch_artist=True)

        # 设置样式
        for box in boxplot['boxes']:
            box.set(facecolor='#1f77b4', alpha=0.7, linewidth=1)

        for median in boxplot['medians']:
            median.set(color='black', linewidth=2)

        # 设置标题和标签 - 使用字体属性
        if chinese_font:
            ax.set_title(col, fontproperties=chinese_font, fontsize=12, fontweight='bold', pad=10)
            ax.set_ylabel('值', fontproperties=chinese_font, fontsize=10)
        else:
            ax.set_title(col, fontsize=12, fontweight='bold', pad=10)
            ax.set_ylabel('值', fontsize=10)

        ax.grid(True, axis='y', linestyle='--', alpha=0.3)

    # 隐藏多余子图
    for j in range(len(current_columns), 4):
        axes[j].set_visible(False)

    # 调整布局
    plt.tight_layout()

    # 保存图表 - 确保字体嵌入
    plt.rcParams['pdf.fonttype'] = 42
    plt.rcParams['ps.fonttype'] = 42
    plt.rcParams['svg.fonttype'] = 'path'

    # 为每个图表生成不同的文件名
    filename = f'solve_1_分面箱线图_{fig_idx+1}.png'
    plt.savefig(filename, dpi=600, bbox_inches='tight')

    print(f"图表已保存为 '{filename}' ")
    plt.close(fig)  # 关闭当前图形，避免内存泄漏

print("所有图表已生成完毕！")
```

```python
"""repeated measure correlation coefficient"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

df_clean = pd.read_csv('data_boy_clean_all.csv', encoding='gbk')

measurement_counts = df_clean['孕妇代码'].value_counts()
print("每个孕妇的测量次数统计:")
print(measurement_counts.describe())

# 只保留有多次测量的孕妇
valid_subjects = measurement_counts[measurement_counts >= 1].index
df_filtered = df_clean[df_clean['孕妇代码'].isin(valid_subjects)]

print(f"\n有效孕妇数量: {len(valid_subjects)}")
print(f"总测量次数: {len(df_filtered)}")

# 实现多变量重复测量分析函数
def multivariate_rmcorr_analysis(data, subject_col, x_cols, y_col):
    """
    执行多变量重复测量相关分析

    参数:
    data: 包含数据的DataFrame
    subject_col: 标识受试者的列名
    x_cols: 自变量列名列表
    y_col: 因变量列名

    返回:
    混合效应模型结果，以及可视化图表
    """
    # 构建混合效应模型公式
    formula = f"{y_col} ~ {' + '.join(x_cols)}"

    # 使用混合效应模型
    model = smf.mixedlm(formula, data, groups=data[subject_col])
    result = model.fit()

    return result, data

# 执行多变量分析
x_cols = ['孕妇BMI', '检测孕周_周数']
y_col = 'Y染色体浓度'

result, df_result = multivariate_rmcorr_analysis(df_filtered, '孕妇代码', x_cols, y_col)

print("\n多变量重复测量分析结果:")
print(result.summary())

# 提取关键结果
print("\n关键统计结果:")
for x in x_cols:
    coeff = result.params[x]
    p_value = result.pvalues[x]
    print(f"{x}: 系数 = {coeff:.30f}, p值 = {p_value:.40f}")

# 可视化每个自变量与因变量的关系
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

for i, x_col in enumerate(x_cols):
    # 计算每个受试者内部的中心化变量
    df_result[f'{x_col}_centered'] = df_result.groupby('孕妇代码')[x_col].transform(lambda x: x - x.mean())
    df_result[f'{y_col}_centered'] = df_result.groupby('孕妇代码')[y_col].transform(lambda x: x - x.mean())

    # 计算个体内相关系数
    r_within = np.corrcoef(df_result[f'{x_col}_centered'], df_result[f'{y_col}_centered'])[0, 1]

    # 绘制散点图
    subjects = df_result['孕妇代码'].unique()
    colors = plt.cm.tab10(np.linspace(0, 1, len(subjects)))

    # 随机选择部分孕妇显示在图例中（避免图例过于拥挤）
    legend_subjects = np.random.choice(subjects, min(8, len(subjects)), replace=False)

    for j, subject in enumerate(subjects):
        subject_data = df_result[df_result['孕妇代码'] == subject]
        color = colors[j % len(colors)]

        # 绘制散点
        axes[i].scatter(subject_data[x_col], subject_data[y_col],
                       color=color, alpha=0.7, s=50)

        # 修改后的绘制个体回归线代码
        if len(subject_data) > 1:
            # 检查x值是否有变异
            if subject_data[x_col].nunique() > 1:  # 确保x值有变化
                slope, intercept, _, _, _ = stats.linregress(subject_data[x_col], subject_data[y_col])
                x_range = np.linspace(subject_data[x_col].min(), subject_data[x_col].max(), 100)

                # 只为图例中的孕妇添加标签
                if subject in legend_subjects:
                    axes[i].plot(x_range, intercept + slope * x_range,
                                color=color, linestyle='--', alpha=0.7, label=f'孕妇 {subject}')
                else:
                    axes[i].plot(x_range, intercept + slope * x_range,
                                color=color, linestyle='--', alpha=0.7)
            else:
                # 如果x值完全相同，绘制水平线
                y_mean = subject_data[y_col].mean()
                axes[i].axhline(y=y_mean, color=color, linestyle='--', alpha=0.7)

    # 绘制总体回归线
    slope, intercept, _, _, _ = stats.linregress(df_result[x_col], df_result[y_col])
    x_range = np.linspace(df_result[x_col].min(), df_result[x_col].max(), 100)
    axes[i].plot(x_range, intercept + slope * x_range, 'k-', linewidth=3, label='总体趋势')

    axes[i].set_xlabel(x_col)
    axes[i].set_ylabel(y_col)
    axes[i].set_title(f'{x_col} vs {y_col}\n个体内相关系数: {r_within:.3f}')
    axes[i].legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.savefig('solve_1_多变量_重复测量分析.png', dpi=600, bbox_inches='tight')
plt.show()

# 绘制残差图检查模型假设
plt.figure(figsize=(10, 6))
df_result['预测值'] = result.predict()
df_result['残差'] = df_result[y_col] - df_result['预测值']

plt.scatter(df_result['预测值'], df_result['残差'], alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('预测值')
plt.ylabel('残差')
plt.title('残差图')
plt.tight_layout()
plt.savefig('solve_1_残差图.png', dpi=600)
plt.show()

# 检查残差的正态性
plt.figure(figsize=(10, 6))
stats.probplot(df_result['残差'], dist="norm", plot=plt)
plt.title('残差Q-Q图')
plt.xlabel('理论分位数')
plt.ylabel('标准化后残差值')
plt.tight_layout()
plt.savefig('solve_1_残差Q-Q图.png', dpi=600)
plt.show()

# 创建3D散点图展示三个变量之间的关系
from matplotlib.lines import Line2D

# 为每个孕妇分配唯一的颜色
subjects = df_result['孕妇代码'].unique()
color_map = {}
colors = plt.cm.tab10(np.linspace(0, 1, len(subjects)))

for i, subject in enumerate(subjects):
    color_map[subject] = colors[i % len(colors)]

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# 随机选择部分孕妇显示在图例中
legend_subjects = np.random.choice(subjects, min(10, len(subjects)), replace=False)

# 创建图例句柄和标签列表
legend_handles = []
legend_labels = []

for subject in subjects:
    subject_data = df_result[df_result['孕妇代码'] == subject]
    color = color_map[subject]

    # 绘制散点
    scatter = ax.scatter(subject_data['孕妇BMI'], subject_data['检测孕周_周数'], subject_data['Y染色体浓度'],
                        color=color, alpha=0.7, s=50)

    # 如果这个孕妇需要在图例中显示，则创建对应的图例句柄
    if subject in legend_subjects:
        proxy_artist = Line2D([0], [0], linestyle='none', marker='o',
                             markersize=8, markerfacecolor=color, markeredgecolor=color)
        legend_handles.append(proxy_artist)
        legend_labels.append(f'孕妇 {subject}')

ax.set_xlabel('孕妇BMI')
ax.set_ylabel('检测孕周(周数)')
ax.set_zlabel('Y染色体浓度')
ax.set_title('孕妇BMI、检测孕周天数与Y染色体浓度的关系')

# 添加图例，使用我们自定义的句柄和标签
ax.legend(legend_handles, legend_labels, bbox_to_anchor=(1.1, 1), loc='upper left')

plt.tight_layout()
plt.savefig('solve_1_3D散点图.png', dpi=600, bbox_inches='tight')
plt.show()
```
